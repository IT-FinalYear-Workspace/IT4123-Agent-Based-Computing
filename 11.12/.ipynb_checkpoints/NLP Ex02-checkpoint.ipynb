```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents=[
    "Natural Language Processing is a subfield pf artificail intelligence",
    "It focuses on the interaction between computers and humans using natural language.",
    "TF-IDF is a technique used to evaluate the importance of words in a dcoument"
]
```

```python
vectoizer=TfidfVectorizer()
tfidf_matrix=vectoizer.fit_transform(documents)
```

```python
feature_names=vectoizer.get_feature_names_out()
print("Feature Names(Vocabulary):")
print(feature_names)
```

```python
idf_values=vectoizer.idf_
print("IDF Values:")
print(idf_values)
```

```python
word_indexes=vectoizer.vocabulary_
print("Word Indexes:")
print(word_indexes)
```

```python
print("TF-IDF Matrix as Sparse Matrix:")
print( tfidf_matrix)
```

```python
dense_matrix=tfidf_matrix.toarray()
print("TF-IDF Matrix as Array:")
print(dense_matrix)
```

```python
for i, document in enumerate(dense_matrix):
    print(f"\nDocument {i+1}:")
    for word,score in zip(feature_names, document):
        if score > 0:
            print(f"{word}: {score:.3f}")
```
