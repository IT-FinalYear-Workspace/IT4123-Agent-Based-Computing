```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
text="Data Science is an exciting field! It combines statistics, programming, and domain knowledge."
print("Original Text:",text)
```

```python
text=text.lower()
print("After Lowercasing:",text)
```

```python
text=text.translate(str.maketrans('', '',string.punctuation))
print("After Removing Punctuation:",text)
```

```python
tokens=word_tokenize(text)
print("After Tokenization:",tokens)
```

```python
stop_words=set(stopwords.words('english'))
filtered_tokens=[word for word in tokens if word not in stop_words]
print("After Removing Stopwords ; ",filtered_tokens)
```

```python
stemmer=PorterStemmer()
stemmed_word=[stemmer.stem(word) for word in filtered_tokens]
print("After Stemming:",stemmed_word)
```

```python
lemmatizer=WordNetLemmatizer()
lemmatized_words=[lemmatizer.lemmatize(word) for word in filtered_tokens]
print("After Lemmatization:",lemmatized_words)
```
